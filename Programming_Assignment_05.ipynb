```python
# 1.Write a Python Program to Find LCM ?
```

```python
def lcm(var1,var2):
    
    if var1 > var2:
        greater = var1
    else:
        greater = var2
        
    while (True):
        if ((greater % var1 == 0) and (greater % var2 == 0)):
            lcm = greater
            break
        greater += 1
    
    return lcm
            
```

```python
lcm(24,15)
```

```python
#2.Write a Python Program to Find HCF?
```

```python
def hcf(var1,var2):
    
    if var1 > var2:
        smaller = var2
    else:
        smaller = var1
        
    for i in range(1,smaller+1):
        if ((var1 % i == 0) and (var2 % i == 0)):
            hcf = i
    
    return hcf
```

```python
hcf(24,36)
```

```python
#5 Write a Python Program to Make a Simple Calculator with 4 basic mathematical operations?
```

```python


def add(a,b):
    return (a+b)

def subtract(a,b):
    return (a-b)

def multiply(a,b):
    return (a*b)

def divide(a,b):
    return (a/b)

print("Calculate four basic calculation")

while True:
    choice = input("Enter opeartion(add/subtract/multiply/divide):  ")
    
    if choice in ("add" , "subtract" , "multiply" , "divide"):
        value_1 = float(input("Enter any number:  "))
        value_2 = float(input("Enter any number:  "))
        
        if choice == "add":
            print(f"Addition is  {add(value_1,value_2)}")
            
        elif choice == "subtract":
            print(f"Subtraction is {subtract(value_1,value_2)}")
            
        elif choice == "multiply":
            print(f"Multiplication is {multiply(value_1,value_2)}")
            
        elif choice == "divide":
            print(f"Division is {divide(value_1,value_2)}")
            
        next_operation = input("Next calculation (yes/no):  ")
        if next_operation == "no":
            break
            
    else:
        print("Invalid input")
```
